```python
import numpy as np
```

```python
w_a = np.load("../raw_data/projections/projected_w_a.npz")["w"]
w_b = np.load("../raw_data/projections/projected_w_a.npz")["w"]
```

```python
import scipy.interpolate

# Define the "time" points for w_a and w_b
x = np.array([0, 1])

# Stack w_a and w_b along the first dimension
y = np.stack([w_a, w_b], axis=0)

# Create the interpolation function
interp = scipy.interpolate.interp1d(x, y, kind='linear', axis=0)

# Define the time points at which you want to interpolate
num_interps = 10
x_interp = np.linspace(0, 1, num_interps)
display('x_interp', x_interp)

# Generate the interpolated w vectors
w_interp = interp(x_interp)
display('w_interp', w_interp)
```
