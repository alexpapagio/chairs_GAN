```python
import os
import shutil

def copy_images_with_pattern(source_folder, destination_folder, pattern):
    # Create destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate through files in the source folder
    for filename in os.listdir(source_folder):
        # Check if the filename contains the pattern
        if pattern in filename:
            # Get the full path of the source file
            source_file = os.path.join(source_folder, filename)
            # Copy the file to the destination folder
            shutil.copy(source_file, destination_folder)
            print(f"Copied {filename} to {destination_folder}")

# Replace these paths with your source and destination folder paths
source_folder = '..//raw_data/archive/rendered_chairs_single_view_100x100_jpg_colour'
destination_folder = '..//raw_data/archive/rendered_chairs_single_view_100x100_jpg_colour_2'
pattern = 'p020'

copy_images_with_pattern(source_folder, destination_folder, pattern)
```
